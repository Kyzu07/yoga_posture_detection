```python
import numpy as np
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras.layers import Dense, Activation
from tensorflow.keras.optimizers import Adam
from tensorflow.keras.metrics import categorical_crossentropy
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from tensorflow.keras.models import Model
from tensorflow.keras.applications import imagenet_utils
from sklearn.metrics import confusion_matrix
import itertools
import os
import shutil
import random
# import splitfolders
import matplotlib.pyplot as plt
from tensorflow.keras.metrics import top_k_categorical_accuracy
from tensorflow.python.keras.utils import np_utils
%matplotlib inline
```

```python
model = tf.keras.models.load_model('/home/kaizu/Downloads/thesis/Finetuned_82_Xception(82%).h5')
```

```python
model.summary()
```

```python
def pred(img):
    test_image = ImageDataGenerator(preprocessing_function=tf.keras.applications.mobilenet.preprocess_input).flow_from_directory(
    directory=img, target_size=(224, 224))
    y_prob = model.predict(x=test_image, verbose=2) 
    y_classes = y_prob.argmax(axis=-1)
    return y_classes[0]
```

```python
train_path = '/media/kaizu/2/Study/Thesis/MobileNet/train'
valid_path = '/media/kaizu/2/Study/Thesis/MobileNet/val'
test_path = '/media/kaizu/2/Study/Thesis/MobileNet/test'
```

```python
train_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.xception.preprocess_input).flow_from_directory(
    directory=train_path, target_size=(224,224), batch_size=10)
valid_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.xception.preprocess_input).flow_from_directory(
    directory=valid_path, target_size=(224,224), batch_size=10)
test_batches = ImageDataGenerator(preprocessing_function=tf.keras.applications.xception.preprocess_input).flow_from_directory(
    directory=test_path, target_size=(224, 224), batch_size=10, shuffle=False)
```

```python
d = test_batches.class_indices

# np.asarray(d.keys())

d = d.keys()

```

```python
from PIL import ImageFile
ImageFile.LOAD_TRUNCATED_IMAGES = True
```

```python
count = 0
list_ = list()
for i in d:
    path = "/media/kaizu/2/Study/Thesis/MobileNet/test/" + str(i) + "/"
    dst_path = "/media/kaizu/2/Study/Thesis/MobileNet/Category/singleImage/"
    if os.path.exists(path):
        l = os.listdir(path)
        for j in l:
            imgPath = path + str(j)
#             print(imgPath)
            shutil.copy(imgPath, dst_path+"folder/")
            p = pred(dst_path)
            list_.append([str(j), (test_batches.class_indices).get(str(i)), p])
            os.remove(dst_path+"folder/"+str(j))
    count += 1

print(count)
    
```
